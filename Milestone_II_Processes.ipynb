#PCA of Final Batch-Norm Imputed BL Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Load the dataset
# final_batch_norm_imputed_BL.csv or final_batch_norm_imputed_V06.csv
df = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/final_batch_norm_imputed_BL.csv")

# Drop metadata columns (assumed to be the first 4)
metadata_cols = df.columns[:4]
feature_cols = df.columns[4:-1]  # assuming 'Class' is the last column

# Extract features and class labels
X = df[feature_cols]
y = df["Class"].map({0: "Control", 1: "PD"})  # Map to readable labels

# Standardize features
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# Create DataFrame for visualization
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Group'] = y.values

# Plot PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Group', s=60)
plt.title('PCA of BL Metabolomics Data')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.2f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.2f}%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Group-Level PCA Centroid Shift with Shared Columns (BL to V06)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np

# Load filtered datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Clean up class labels
df_bl["Group"] = df_bl["Class"].map({0: "Control", 1: "PD"})
df_v06["Group"] = df_v06["Class"].map({0: "Control", 1: "PD"})

# Extract features (exclude metadata)
feature_cols = [col for col in df_bl.columns if col not in ["CLIENT_IDENTIFIER", "Class", "Group"]]
X_bl = df_bl[feature_cols]
X_v06 = df_v06[feature_cols]

# Standardize using the combined dataset to keep scales consistent
X_all = pd.concat([X_bl, X_v06])
X_scaled = StandardScaler().fit_transform(X_all)

# Split back
X_bl_scaled = X_scaled[:len(X_bl)]
X_v06_scaled = X_scaled[len(X_bl):]

# Apply PCA (on entire set to get shared components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
pca_bl = X_pca[:len(X_bl)]
pca_v06 = X_pca[len(X_bl):]

# Add PCA coordinates back
df_bl["PC1"] = pca_bl[:, 0]
df_bl["PC2"] = pca_bl[:, 1]
df_v06["PC1"] = pca_v06[:, 0]
df_v06["PC2"] = pca_v06[:, 1]

# Compute centroids for each group
centroids_bl = df_bl.groupby("Group")[["PC1", "PC2"]].mean()
centroids_v06 = df_v06.groupby("Group")[["PC1", "PC2"]].mean()

# Plot
plt.figure(figsize=(8, 6))
colors = {"Control": "blue", "PD": "red"}

for group in ["Control", "PD"]:
    x0, y0 = centroids_bl.loc[group]
    x1, y1 = centroids_v06.loc[group]

    plt.scatter(x0, y0, color=colors[group], label=f"{group} BL", s=100, edgecolor='black')
    plt.scatter(x1, y1, color=colors[group], marker="X", label=f"{group} V06", s=100)
    plt.arrow(x0, y0, x1 - x0, y1 - y0, color=colors[group], alpha=0.6,
              head_width=0.3, length_includes_head=True)

plt.title("Group-Level PCA Centroid Shift (BL to V06)")
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.2f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.2f}%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Volcano Plot Code (PD BL vs. PD V06)

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Load filtered datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Filter to PD patients only
df_bl_pd = df_bl[df_bl["Class"] == 1]
df_v06_pd = df_v06[df_v06["Class"] == 1]

# Get shared feature columns
feature_cols = [col for col in df_bl.columns if col not in ["CLIENT_IDENTIFIER", "Class"]]

# Initialize results
results = []

for col in feature_cols:
    # t-test
    t_stat, p_val = ttest_ind(df_v06_pd[col], df_bl_pd[col], equal_var=False, nan_policy='omit')

    # log2 fold change: V06 vs BL
    mean_bl = df_bl_pd[col].mean()
    mean_v06 = df_v06_pd[col].mean()
    log2_fc = np.log2(mean_v06 + 1e-8) - np.log2(mean_bl + 1e-8)  # Add small value to avoid log(0)

    results.append({
        "metabolite": col,
        "log2FC": log2_fc,
        "p_value": p_val
    })

# Convert to DataFrame
volcano_df = pd.DataFrame(results)
volcano_df["-log10(p)"] = -np.log10(volcano_df["p_value"])

# Add significance labels
p_thresh = 0.05
fc_thresh = 1  # log2 fold change threshold
volcano_df["significance"] = "Not Significant"
volcano_df.loc[(volcano_df["p_value"] < p_thresh) & (volcano_df["log2FC"] > fc_thresh), "significance"] = "Up in V06"
volcano_df.loc[(volcano_df["p_value"] < p_thresh) & (volcano_df["log2FC"] < -fc_thresh), "significance"] = "Down in V06"

# Plot
plt.figure(figsize=(10, 7))
sns.scatterplot(data=volcano_df, x="log2FC", y="-log10(p)", hue="significance", style="significance",
                palette={"Up in V06": "red", "Down in V06": "blue", "Not Significant": "gray"}, alpha=0.7)

plt.axhline(-np.log10(p_thresh), linestyle='--', color='black', linewidth=1)
plt.axvline(fc_thresh, linestyle='--', color='black', linewidth=1)
plt.axvline(-fc_thresh, linestyle='--', color='black', linewidth=1)

plt.title("Volcano Plot: PD Baseline vs. 6-Month Follow-Up")
plt.xlabel("Log2 Fold Change (V06 vs BL)")
plt.ylabel("-Log10(p-value)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Code to List Top Significant Metabolites

```python
# Filter for significant hits
sig_hits = volcano_df[
    (volcano_df["p_value"] < 0.05) & (abs(volcano_df["log2FC"]) >= 1)
].copy()

# Sort by p-value
sig_hits = sig_hits.sort_values("p_value")

# View top hits
print("Top significant metabolites (sorted by p-value):")
print(sig_hits[["metabolite", "log2FC", "p_value"]].head(10))
```

#XGBoost Classifier Code (Baseline Only)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# Load baseline data
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")

# Prepare features and labels
X = df_bl.drop(columns=["CLIENT_IDENTIFIER", "Class"])
y = df_bl["Class"]  # 0 = Control, 1 = PD

# 56 were used to train the model
# 14 were used to test the model
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Define and train XGBoost model
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Control", "PD"], yticklabels=["Control", "PD"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# Feature importance (top 20)
xgb.plot_importance(model, max_num_features=20, importance_type='gain', height=0.5)
plt.title("Top 20 Metabolites by Importance (XGBoost)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# Load baseline data
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Prepare features and labels
X = df_bl.drop(columns=["CLIENT_IDENTIFIER", "Class"])
y = df_bl["Class"]  # 0 = Control, 1 = PD

# 56 were used to train the model
# 14 were used to test the model
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Define and train XGBoost model
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Control", "PD"], yticklabels=["Control", "PD"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# Feature importance (top 20)
xgb.plot_importance(model, max_num_features=20, importance_type='gain', height=0.5)
plt.title("Top 20 Metabolites by Importance (XGBoost)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Map group labels
df_bl["Group"] = df_bl["Class"].map({0: "Control", 1: "PD"})
df_v06["Group"] = df_v06["Class"].map({0: "Control", 1: "PD"})

# Define top metabolite IDs and their names
top_metab_ids = ["100001181", "100000265", "100019968", "100001613", "100001086"]
top_metab_names = {
    "100001181": "docosapentaenoate (22:5n3)",
    "100000265": "kynurenine",
    "100019968": "lyxonate",
    "100001613": "tetradecanedioate (C14-DC)",
    "100001086": "N-(2-furoyl)glycine"
}
combined_labels = {k: f"{v}\n[{k}]" for k, v in top_metab_names.items()}

# Prepare baseline data
df_plot_bl = df_bl[["Group"] + top_metab_ids].rename(columns={k: combined_labels[k] for k in top_metab_ids})
df_melted_bl = df_plot_bl.melt(id_vars="Group", var_name="Metabolite", value_name="Abundance")

# Prepare V06 data
df_plot_v06 = df_v06[["Group"] + top_metab_ids].rename(columns={k: combined_labels[k] for k in top_metab_ids})
df_melted_v06 = df_plot_v06.melt(id_vars="Group", var_name="Metabolite", value_name="Abundance")

def plot_boxplot_with_means(df_melted, title):
    plt.figure(figsize=(12, 6))
    ax = sns.boxplot(data=df_melted, x="Metabolite", y="Abundance", hue="Group", showfliers=False)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")

    # Calculate means
    group_means = df_melted.groupby(["Metabolite", "Group"])["Abundance"].mean().reset_index()

    # Add labels above boxes
    for i, box in enumerate(ax.artists):
        metabolite_labels = [t.get_text() for t in ax.get_xticklabels()]
        m_idx = i % len(metabolite_labels)
        metabolite = metabolite_labels[m_idx]
        group = "PD" if i >= len(metabolite_labels) else "Control"
        mean_val = group_means.query("Metabolite == @metabolite and Group == @group")["Abundance"].values[0]
        x = m_idx + (-0.2 if group == "Control" else 0.2)
        ax.text(x, mean_val + 0.05, f"{mean_val:.2f}", ha='center', va='bottom', fontsize=9, color='black')

    plt.tight_layout()
    plt.show()

# Plot for Baseline
plot_boxplot_with_means(df_melted_bl, "Top XGBoost Metabolites: PD vs Control (Baseline)\n[Means Labeled, Outliers Hidden]")

# Plot for V06
plot_boxplot_with_means(df_melted_v06, "Top XGBoost Metabolites: PD vs Control (6-Month Follow-Up)\n[Means Labeled, Outliers Hidden]")
```

#Random Forest on Baseline Data

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load baseline dataset
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")

# Prepare data
X = df_bl.drop(columns=["CLIENT_IDENTIFIER", "Class"])
y = df_bl["Class"]  # 0 = Control, 1 = PD

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf_model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Control", "PD"], yticklabels=["Control", "PD"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Random Forest Confusion Matrix BL")
plt.tight_layout()
plt.show()

# Feature importance plot (top 20)
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1][:20]
top_features = X.columns[indices]
top_importances = importances[indices]

plt.figure(figsize=(10, 6))
sns.barplot(x=top_importances, y=top_features, color='green')
plt.title("Top 20 Feature Importances BL (Random Forest)")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```

#Combine and Train Random Forest

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Add timepoint label
df_bl["Timepoint"] = "BL"
df_v06["Timepoint"] = "V06"

# Keep only shared columns (for safety)
shared_cols = list(set(df_bl.columns) & set(df_v06.columns))
shared_cols = [c for c in shared_cols if c not in ["CLIENT_IDENTIFIER", "Class", "Timepoint"]]

# Reconstruct consistent dataframes
df_bl = df_bl[["Class", "Timepoint"] + shared_cols]
df_v06 = df_v06[["Class", "Timepoint"] + shared_cols]

# Combine datasets
df_all = pd.concat([df_bl, df_v06], ignore_index=True)

# One-hot encode timepoint
df_all = pd.get_dummies(df_all, columns=["Timepoint"], drop_first=True)  # Only keeps "Timepoint_V06"

# Prepare input/output
X = df_all.drop(columns=["Class"])
y = df_all["Class"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Control", "PD"], yticklabels=["Control", "PD"])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Random Forest (BL + V06 Combined)")
plt.tight_layout()
plt.show()

# Feature importance
importances = model.feature_importances_
indices = np.argsort(importances)[::-1][:20]
top_features = X.columns[indices]
top_importances = importances[indices]

plt.figure(figsize=(10, 6))
sns.barplot(x=top_importances, y=top_features, color='green')
plt.title("Top 20 Feature Importances (Combined Timepoints)")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```

#Cross-Validation Code (Random Forest, Combined Timepoints)

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_validate
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score

# Load datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")
df_bl["Timepoint"] = "BL"
df_v06["Timepoint"] = "V06"

# Get shared features
shared_cols = list(set(df_bl.columns) & set(df_v06.columns))
shared_cols = [c for c in shared_cols if c not in ["CLIENT_IDENTIFIER", "Class", "Timepoint"]]

# Combine datasets
df_bl = df_bl[["Class", "Timepoint"] + shared_cols]
df_v06 = df_v06[["Class", "Timepoint"] + shared_cols]
df_all = pd.concat([df_bl, df_v06], ignore_index=True)

# One-hot encode timepoint
df_all = pd.get_dummies(df_all, columns=["Timepoint"], drop_first=True)

# Features and labels
X = df_all.drop(columns=["Class"])
y = df_all["Class"]

# Define model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Define cross-validation strategy and scoring metrics
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}

# Perform cross-validation
cv_results = cross_validate(rf_model, X, y, cv=cv, scoring=scoring)

# Display results
for metric in scoring.keys():
    scores = cv_results[f'test_{metric}']
    print(f"{metric.capitalize()} (mean ± std): {scores.mean():.3f} ± {scores.std():.3f}")
```

#Stable Feature Importance Across Folds

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

# Load datasets
df_bl = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_BL.csv")
df_v06 = pd.read_csv("/content/drive/MyDrive/CSC-587/Milestone/Data/fbni_shared_columns_V06.csv")

# Add timepoint column
df_bl["Timepoint"] = "BL"
df_v06["Timepoint"] = "V06"

# Keep only shared columns
shared_cols = list(set(df_bl.columns) & set(df_v06.columns))
shared_cols = [c for c in shared_cols if c not in ["CLIENT_IDENTIFIER", "Class", "Timepoint"]]

# Rebuild dataframes with shared columns only
df_bl = df_bl[["Class", "Timepoint"] + shared_cols]
df_v06 = df_v06[["Class", "Timepoint"] + shared_cols]

# Combine datasets
df_all = pd.concat([df_bl, df_v06], ignore_index=True)

# One-hot encode timepoint column
df_all = pd.get_dummies(df_all, columns=["Timepoint"], drop_first=True)

# Separate features and labels
X = df_all.drop(columns=["Class"])
y = df_all["Class"]

# Initialize Random Forest and cross-validation
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Collect feature importances across folds
feature_importance_matrix = []
for train_index, test_index in cv.split(X, y):
    X_train_cv, y_train_cv = X.iloc[train_index], y.iloc[train_index]
    model = clone(rf_model)
    model.fit(X_train_cv, y_train_cv)
    feature_importance_matrix.append(model.feature_importances_)

# Convert to DataFrame
feature_importance_df = pd.DataFrame(feature_importance_matrix, columns=X.columns)

# Compute mean and standard deviation
feature_importance_summary = pd.DataFrame({
    "Feature": X.columns,
    "Mean_Importance": feature_importance_df.mean(),
    "Std_Importance": feature_importance_df.std()
}).sort_values(by="Mean_Importance", ascending=False).reset_index(drop=True)

# Display top 20
print("Top Stable Features Across Folds:")
print(feature_importance_summary.head(20))
```
